```python
import pandas as pd
import docx
import glob
```

```python
##Wybór metody pobierania plików
print("Wybierz metodę importowania tekstu: ")
print("file - pobiera tekst z wszystkich plików docx w tym samym folderze")
print("text - wklejasz tekst ręcznie")
method = input("").lower().strip()

## Jeżeli wolisz wkleić tekst zrób to tutaj
tekst_do_wklejenia = ""




text = ""
if method == "file":
    file_list = glob.glob('*.docx')

    if len(file_list) == 0:
        print(f"Brak plików do pobrania")

    else:
        for file in file_list:
            doc = docx.Document(file)  

            for para in doc.paragraphs:
                text += para.text + " "
        print(f'Sukces! Wczytano dane z {len(file_list)} plików.')

elif method == "texst":
    text = tekst_do_wklejenia
else:
    print("Nieznana metoda! Zrestartuj program.")


print(text)



#Oczyszczanie tekstu
text = text.lower()
text = text.replace(".", "")
text = text.replace(",", "")
text = text.replace("-", "")
text = text.replace("%", "")


print(text)


#Zamiana tekstu na tabelę

text = text.split()
print(text)


#Zliczanie słów
final_table = []
counted = []
for word in text:
    if word not in counted:
        Qty = text.count(word)
        
        
        final_table.append([word, Qty])
        counted.append(word)

#Zamiana na tabelę & wyświetlanie wyników        

final_table.sort(key=lambda x: x[1], reverse=True)
df = pd.DataFrame(final_table, columns=['word', 'Ilosc'])
df
```
