```python
temp_skala = input(f"Podaj w jakiej skali jest temperatura. \nMożesz wpisać C, F lub K. \nMożesz też wpisać pełną nazwę skali: Celsjusz, Farenhait lub Kelwin. \nSkala: ")

if temp_skala in ["C", "Celsjusz"]:
    print(f"Podałeś skalę Celsjusza")
    skala = "C"

elif temp_skala in ["F", "Farenhait"]:
    print(f"Podałeś skalę Fahrenheita")
    skala = "F"

elif temp_skala in ["K", "Kelvin"]:
    print("Podałeś skalę Kelwina")
    skala = "K"

else:
    print(f"Nie podałeś poprawnej skali")    
    print(f"Spróbuj wpisać C, F lub K")
    print(f"Program zostanie zakończony")
    exit()

temp_temperatura = float(input(f"Podaj temperaturę: "))


Celcjusz = 0
Farenhtait = 0
Kelwin = 0

print(f"Podałeś temperaturę: {temp_temperatura} {skala}")
print(f"Temperatura w innych skalach:")

if skala == "C":
    Celcjusz = temp_temperatura
    Farenhtait = (Celcjusz * 9/5) + 32
    Kelwin = Celcjusz + 273.15

    print(f"Farenhait: {round(Farenhtait, 2)}")
    print(f"Kelwin: {round(Kelwin, 2)}")


elif skala == "F":
    Farenhtait = temp_temperatura
    Celcjusz = (Farenhtait - 32) * 5/9
    Kelwin = Celcjusz + 273.15

    print(f"Celsjusz: {round(Celcjusz, 2)}")
    print(f"Kelwin: {round(Kelwin, 2)}")

else:
    Kelwin = temp_temperatura
    Celcjusz = Kelwin - 273.15
    Farenhtait = (Celcjusz * 9/5) + 32

    print(f"Celsjusz: {round(Celcjusz, 2)}")
    print(f"Farenhait: {round(Farenhtait, 2)}")


    
```
